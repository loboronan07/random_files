```python
def displayPathtoPrincess(n,grid):
    center = (int(n/2), int(n/2))
    q = [center]
    visited = [center]
    parent = {center:(None, None)}
    
    while q:
        r, c = q.pop(0)
        
        if grid[r][c] == 'p':
            print_path(parent, (r,c))
            return
        
        if r-1 >= 0 and (r-1, c) not in visited:
            q.append((r-1, c))
            visited.append((r-1, c))
            parent[(r-1, c)] = ((r,c), "UP")
        
        if r+1 < n and (r+1, c) not in visited:
            q.append((r+1, c))
            visited.append((r+1, c))
            parent[(r+1, c)] = ((r,c), "DOWN")
        
        if c-1 >= 0 and (r, c-1) not in visited:
            q.append((r, c-1))
            visited.append((r, c-1))
            parent[(r, c-1)] = ((r,c), "LEFT")
        
        if c+1 < n and (r, c+1) not in visited:
            q.append((r, c+1))
            visited.append((r, c+1))
            parent[(r, c+1)] = ((r,c), "RIGHT")
            
def print_path(parent, curr):
    if parent[curr][0]:
        print_path(parent, parent[curr][0])
        print(parent[curr][1])
    

m = int(input())
grid = [] 
for i in range(0, m): 
    grid.append(input().strip())

displayPathtoPrincess(m,grid)
```

```python
def heuristic(x, y, r, c):
    return abs(x-r) + abs(y-c)

def getPathToFood(x, y, r, c, m, n, grid):
    visited = []
    q = [(x, y), heuristic(x, y, r, c)]
    parent = {(x,y): (None, None)}
    while q:
        q.sort(key=lambda x:x[1])
        pos, cost = q.pop(0)
        x, y = pos
        h = heuristic(x, y, r, c)
        if h == 0:
            path = []
            gen_path(parent, (x,y), path)
            return path
        
        pathcost = cost - h
        
        if x-1>=0 and grid[x-1][y] != '%' and (x-1, y) not in visited:
            parent[(x-1, y)] = ((x,y), "UP")
            q.append(((x-1, y), pathcost+heuristic(x-1, y, r, c))
        
        if y-1>=0 and grid[x][y-1] != '%' and (x, y-1) not in visited:
            parent[(x, y-1)] = ((x,y), "LEFT")
            q.append(((x, y-1), pathcost+heuristic(x, y-1, r, c))
        
        if y+1<n and grid[x][y+1] != '%' and (x, y+1) not in visited:
            parent[(x, y+1)] = ((x,y), "RIGHT")
            q.append(((x, y+1), pathcost+heuristic(x, y+1, r, c))
        
        if x+1<m and grid[x+1][y] != '%' and (x+1, y) not in visited:
            parent[(x+1, y)] = ((x,y), "DOWN")
            q.append(((x+1, y), pathcost+heuristic(x+1, y, r, c))
        
def gen_path(parent, curr, path):
    if parent[curr][0]:
        gen_path(parent, parent[curr][0], path)
        path.append(parent[curr][1])

if __name__ == "__main__":
    pos_pac = [int(i) for i in input().strip().split()]
    pos_food = [int(i) for i in input().strip().split()]
    size = [int(i) for i in input().strip().split()]
    grid = [[j for j in input().strip()] for i in range(size[0])]
    
    path = getPathToFood(pos_pac[0], pos_pac[1], pos_food[0], pos_food[1], size[0], size[1], grid)
    
    for move in path:
        print(move)
```
